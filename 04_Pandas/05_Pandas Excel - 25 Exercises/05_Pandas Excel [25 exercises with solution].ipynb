### 7. Write a Pandas program to add summation to a row of the given excel file.

```python
df = pd.read_excel('coalpublic2013.xlsx', header = 3)
sum_row = df[['Production (short tons)', 'Labor Hours']].sum()
df_sum = pd.DataFrame(data = sum_row).T      # .T denotes transpose
df_sum = df_sum.reindex(columns = df.columns)
df_sum
```
